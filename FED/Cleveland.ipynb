## 1.1 Cleveland FED Consumers and Covid 19

- [Data](https://www.clevelandfed.org/our-research/indicators-and-data/consumers-and-covid-19.aspx)



```python
url = "https://www.clevelandfed.org/~/media/files/webcharts/consumersandcovid19/consumers_and_covid19.xlsx?la=en"
df = pd.read_excel(url, index_col = 0, parse_dates=True)
df.head(2)
```

```python
df.columns
```

```python
df.loc[:, 'Expected coronavirus impact on GDP, next 12 months, robust mean, 7-day avg'].interpolate().plot()
plt.title('Expected coronavirus impact on GDP, next 12 months')
#plt.legend(bbox_to_anchor=[0.5, -.3], loc='lower center', ncol=4, shadow=True)
plt.ylabel('y-o-y pct')
plt.xlabel('')

txt="SRC: Federal Reserve Bank of Cleveland"
plt.figtext(.15, .01, txt, fontweight='bold', wrap=True, horizontalalignment='center', fontsize=12);
```
